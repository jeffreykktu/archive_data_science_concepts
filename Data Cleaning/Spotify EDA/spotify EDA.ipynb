```python
# Cleaning the genre column
# since the more generic genre is at the back of the list, extract the last 1 genre
 
genre_1_list = [] #, genre_2_list, genre_3_list = [], [], []
for genres in df_edit.spotify_genre.values:
    genre_str_list = str(genres).replace("['", "").replace("']","").replace("'","").split(",")
    genre_1_list.append(genre_str_list[-1])

print(genre_1_list)
len(genre_1_list)
```

```python
df_edit.reset_index(drop=True, inplace=True)
```

```python
# replace the genre column with that only has 1 genre
df_edit['genre'] = pd.Series(genre_1_list, index=df_edit.index)
df_edit = df_edit.drop("spotify_genre", axis="columns")
df_edit
```

```python
# Drop Non useful info for Regression Analysis
df_edit2 = df_edit.drop(['SongID', 'spotify_track_id'], axis='columns')
```

### Fill in the null values for all the measurements

```python
df_edit2['genre'] = df_edit2['genre'].apply(lambda x: x.replace("[]", "").replace("]","").replace('"', '').replace("[","").strip())
df_edit2['genre'] = df_edit2['genre'].apply(lambda x: x.replace("NA", "nan"))
df_edit2['genre'].value_counts()
df_edit2['genre'].unique()
```

```python
# impute null values for all the measurements by filling the means of those measurements group by genres

cols = ["danceability", "energy", "key", "loudness", "mode", "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "time_signature"]
# len(cols)
df_genres_means = pd.DataFrame()
for col in cols:
    df_genres_means = pd.concat([df_genres_means, df_edit2.groupby("genre").agg({col:"mean"})], axis=1)

df_genres_means
```

```python
df_edit2.genre[df_edit2['danceability'].isnull()].value_counts()
df_edit2.genre[df_edit2['energy'].isnull()].value_counts()
df_edit2.genre[df_edit2['loudness'].isnull()].value_counts()
```

```python
# Impute the null value of features in southern hip hop with the mean

mean_cols = ["danceability", "tempo", "energy", "loudness", "speechiness", "acousticness", "instrumentalness", "liveness", "valence"]
discrete_cols = ["key", "time_signature", "mode"]
feature_vals = {}
for col in mean_cols:
    feature_vals[col] = df_genres_means.loc['southern hip hop', col]
print(feature_vals)
df_edit2[df_edit2.genre=='southern hip hop']=df_edit2[df_edit2.genre=='southern hip hop'].fillna(value=feature_vals)

```

```python
# Handle southern hip hop's time signature and mode
df_edit2[df_edit2.genre=='southern hip hop'] = df_edit2[df_edit2.genre=='southern hip hop'].fillna(value={"time_signature":4.0, "mode":1.0})
```

```python
southern_key_median=df_edit2[df_edit2.genre=='southern hip hop']['key'].median()
```

```python
df_edit2[df_edit2.genre=='southern hip hop'] = df_edit2[df_edit2.genre=='southern hip hop'].fillna(value={"key": southern_key_median})
```

```python
df_edit2[df_edit2.genre=='southern hip hop'].isnull().sum()
```
